```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import statistics

df1 = pd.read_csv('/content/student_bigdata.csv')
print(df1.head())    #first 5 rows get displayed
print()

datashape = df1.shape
print(datashape)
print()

print(df1.columns)
print()

num_missing = df1.isna().sum()
print(num_missing)

#mean:
li = list(df1.columns)
li_dash = []
means = []
means.append(np.mean(df1[li[0]]))
li_dash.append(li[0])
means.append(np.mean(df1[li[2]]))
li_dash.append(li[2])
means.append(np.mean(df1[li[3]]))
means.append(np.mean(df1[li[4]]))
means.append(np.mean(df1[li[5]]))
means.append(np.mean(df1[li[6]]))
means.append(np.mean(df1[li[7]]))
li_dash.append(li[3])
li_dash.append(li[4])
li_dash.append(li[5])
li_dash.append(li[6])
li_dash.append(li[7])
print()

for x, y in zip(means, li_dash):
    print(y, x)
print()

#median:
medians = []
medians.append(np.median(df1[li[0]]))
medians.append(np.median(df1[li[2]]))
medians.append(np.median(df1[li[3]]))
medians.append(np.median(df1[li[4]]))
medians.append(np.median(df1[li[5]]))
medians.append(np.median(df1[li[6]]))
medians.append(np.median(df1[li[7]]))

for x, y in zip(medians, li_dash):
  print(y, x)
print()

#std deviation:
std_devs = []
std_devs.append(np.std(df1[li[0]]))
std_devs.append(np.std(df1[li[2]]))
std_devs.append(np.std(df1[li[3]]))
std_devs.append(np.std(df1[li[4]]))
std_devs.append(np.std(df1[li[5]]))
std_devs.append(np.std(df1[li[6]]))
std_devs.append(np.std(df1[li[7]]))

for x, y in zip(std_devs, li_dash):
  print(y, x)

#plots:
print()
sns.histplot(df1[li[4]], kde = True)

```
